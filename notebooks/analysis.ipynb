```python
# Converting `ORDERDATE` to datetime

df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df.info()
```

### VALIDATING SALES

```python
df['CALCULATED_SALES'] = df['QUANTITYORDERED'] * df['PRICEEACH']
df.round()  #used `.round()` to round 1 decimal places.
```

### OBSERVATION
