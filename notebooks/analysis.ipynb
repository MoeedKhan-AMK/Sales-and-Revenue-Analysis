### DATA CLEANING

```python
# Converting `ORDERDATE` to datetime

df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df.info()
```

### VALIDATING SALES

```python
df['CALCULATED_SALES'] = df['QUANTITYORDERED'] * df['PRICEEACH']
df.round()  #used `.round()` to round 1 decimal places.
```

```python
df.describe()
```

### OBSERVATION
- New Column `Calculated_sales` create to validate sales column
- `Sales` columns seems not reliable.
- For Example: `QUANTITYORDERED: 41 * PRICEEACH: 100 == 4100`, but in `SALES` column it was `4708.0`

### EXPLORATORY DATA ANALYSIS (EDA)
- Total sales per year
- Total sales per month

```python
df_month_sales = df.groupby('MONTH_ID')['CALCULATED_SALES'].sum()
```

```python
df_year_sales = df.groupby('YEAR_ID')['CALCULATED_SALES'].sum()
df_year_sales
```

##### CUSTOMER ORDER PERFORMANCE 
- Average order value (AOV) = Revenue/number of orders
- Largest single order
- Smallest non-zero order
- Identify Top 5 highest revenue orders.

```python
#AOV
df_aov = df.groupby('ORDERNUMBER')['CALCULATED_SALES'].sum().mean()
df_aov
```

```python
#Largest Single Order
df_largest_single_order = df.loc[df['CALCULATED_SALES'].idxmax()]
df_largest_single_order
```

```python
# Smallest Non Zero Order
df_smallest_non_zero_order = df[df['CALCULATED_SALES'] > 0].loc[df[df['CALCULATED_SALES'] > 0]['CALCULATED_SALES'].idxmin()]
df_smallest_non_zero_order
```

```python
#highest revenue orders
df_highest_revenue_order = df.groupby('ORDERNUMBER')['CALCULATED_SALES'].sum()

highest_order_id = df_highest_revenue_order.idxmax()
highest_order_value = df_highest_revenue_order.max()

highest_order_id, highest_order_value
```

### VISUALIZATION

```python
#PLOTING MONTHLY SALES TREND

plt.figure(figsize=(8,5))
plt.plot(df_month_sales.index, df_month_sales.values)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.show()
```

```python
#PLOTING YEARLY SALES TREND

plt.figure(figsize=(8,5))
plt.plot(df_year_sales.index, df_year_sales.values)

plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")

plt.show()
```

```python
#TOP PRODUCTS BY REVENUE
top_products = df.groupby('PRODUCTLINE')['CALCULATED_SALES'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
plt.bar(top_products.index, top_products.values)

plt.title("Top 10 Products by Revenue")
plt.xlabel("Product Line")
plt.ylabel("Revenue")

plt.show()
```
